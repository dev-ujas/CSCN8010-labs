```python
from sklearn import linear_model
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_house = pd.read_csv("./datasets/housing.csv")
df_house.head()
```

Filtering data baseon 'ocean_proximity'

```python
df_near_bay = df_house[["latitude","longitude","ocean_proximity"]][df_house["ocean_proximity"] == "NEAR BAY"]
print(df_near_bay.head())
df_inland = df_house[["latitude","longitude","ocean_proximity"]][df_house["ocean_proximity"] == "INLAND"]
print(df_inland.head())
df_one_hour = df_house[["latitude","longitude","ocean_proximity"]][df_house["ocean_proximity"] == "<1H OCEAN"]
print(df_one_hour.head())
```

```python
plt.xlabel = "Latitude"
plt.ylabel = "Longitude"
plt.scatter(df_near_bay['latitude'], df_near_bay['longitude'], marker="^",alpha=0.3, label = "near_bay")
plt.scatter(df_inland['latitude'], df_inland['longitude'], marker="+",alpha=0.3, label = "inland")
plt.scatter(df_one_hour['latitude'], df_one_hour['longitude'], marker="*",alpha=0.3, label = "< hour")
plt.legend()
```
